# Data Preprocessing

```python
#Separate Features and Target

x = df.drop('Class', axis=1) #features
y = df['Class'] #Target label(fraud/Not-fraud)
```

```python
#Train-Test Split (80% train, 20% test)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
```

```python
#Scale the ‘Amount’ feature

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train['Amount'] = scaler.fit_transform(X_train[['Amount']])
X_test['Amount'] = scaler.transform(X_test[['Amount']])
```

```python
#SMOTE
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)

# Apply only to training data
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

print('\n Before smote:',y_train.value_counts())
print('\n After smote:',y_train_smote.value_counts())
```

```python
#LOGISTIC REGRESSION

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, average_precision_score

log_reg = LogisticRegression(max_iter=5000, class_weight='balanced', random_state=42)
log_reg.fit(X_train, y_train)

y_pred_lr = log_reg.predict(X_test)
y_prob_lr = log_reg.predict_proba(X_test)[:,1]

print('\n classification_report:\n',classification_report(y_test,y_pred_lr))
print('\n confusion_matrix:\n',confusion_matrix(y_test,y_pred_lr))

print('\n roc_auc_score:\n',roc_auc_score(y_test,y_prob_lr))
print('\n AUPRC (Precision-Recall AUC):\n',average_precision_score(y_test,y_prob_lr))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=300, max_depth=None, class_weight='balanced',random_state=42, n_jobs=-1)

rf.fit(X_train_smote, y_train_smote)

y_pred_rf = rf.predict(X_test)
y_prob_rf = rf.predict_proba(X_test)[:,1]

print("\n=== RANDOM FOREST RESULTS ===")
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))
print("\nClassification Report:\n", classification_report(y_test, y_pred_rf))
print("ROC-AUC:", roc_auc_score(y_test, y_prob_rf))
print("AUPRC:", average_precision_score(y_test, y_prob_rf))
```

```python
from xgboost import XGBClassifier

# Handle imbalance properly (no need SMOTE for boosting models)
fraud_ratio = y_train.value_counts()[0] / y_train.value_counts()[1]

xgb = XGBClassifier(n_estimators=250, learning_rate=0.05, max_depth=4, subsample=0.8, colsample_bytree=0.8, scale_pos_weight=fraud_ratio,
                    eval_metric='logloss', tree_method='hist', n_jobs=-1, random_state=42)

# Using original X_train (NOT SMOTE) because boosting handles imbalance already
xgb.fit(X_train, y_train)

y_pred_xgb = xgb.predict(X_test)
y_prob_xgb = xgb.predict_proba(X_test)[:, 1]

print("\n=== XGBOOST RESULTS ===")
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_xgb))
print("\nClassification Report:\n", classification_report(y_test, y_pred_xgb))
print("ROC-AUC:", roc_auc_score(y_test, y_prob_xgb))
print("AUPRC:", average_precision_score(y_test, y_prob_xgb))
```

```python
y_prob_rf = rf.predict_proba(X_test)[:,1]
```

```python
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt

precision, recall, thresholds = precision_recall_curve(y_test, y_prob_rf)

plt.figure(figsize=(6,4))
plt.plot(thresholds, precision[:-1], label='Precision')
plt.plot(thresholds, recall[:-1], label='Recall')
plt.xlabel('Threshold')
plt.ylabel('Score')
plt.title('Precision & Recall vs Threshold')
plt.legend()
plt.show()
```

```python
threshold = 0.50  # optimal from PR curve

y_pred_final = (y_prob_rf >= threshold).astype(int)

print(confusion_matrix(y_test, y_pred_final))
print(classification_report(y_test, y_pred_final))
```
